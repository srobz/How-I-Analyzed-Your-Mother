# Introduction

The purpose of this project is to do a thorough exploratory data analysis of the characters and scripts of the TV show How I Met Your Mother, as well as create a model that can predict who said which line. The data is from the website https://transcripts.foreverdreaming.org/viewforum.php?f=177 where there are nine pages containing the 208 episodes from nine seasons that the show ran.

This notebook contains the web scraping code, where I scrape the previously mentioned website to get the following information about each episode and created a dataframe to hold it all: Season, Episode, Character, and Line. The process in which I did this involved scraping the nine pages to get a list of the link to each script and a list of the name of each episode with their season/episode number. Then, using a for loop with a timer, I looped through those lists to create a list for Season, Episode, Character, and Line, for each main character's line in each individual episode; these lists were then pushed into a final dataframe.

# Building Dataframe from Scraped Pages

Now that we've scraped the 9 pages and made two lists, one containing all the episode titles and one containing all the links to each script, we'll build a for loop that will parse through each of these lists and create a dataframe containing the season number, episode number, character speaking, and line spoken. This dataframe will only include lines from the five main characters as these are the only characters the final analysis will be on.

```python
season = [] #Making season list
episode = [] #Making episode list
character = [] #Making character list
line = [] #Making line list

start_time = time.time() #Setting start time
seconds = input('Enter: ') #Setting input for seconds
seconds = int(seconds) #Setting seconds to integer to enter

for link in ep_links: #Starting for loop for links in link list
    for i in range(0, 207): #Making it count to 208 episodes so it goes through every link
        html_page = requests.get(ep_links[i]) #Getting script from link list
        soup = BeautifulSoup(html_page.content, 'html.parser') #Making soup read page
        for j in soup.find('div', class_ = 'postbody').findAll('p'): #For loop that removes html and looks for lines
            if 'Ted: ' in j.text: #If line starts with Ted speaking
                season.append(ep_titles[i][:2]) #add season to season list
                episode.append(ep_titles[i][3:][:2]) #add episode to episode list
                character.append('Ted') #add Ted to character list
                line.append(j.text[5:]) #add following line to line list
            elif 'Marshall: ' in j.text: #if line starts with Marshall speaking
                season.append(ep_titles[i][:2]) #add season to season list
                episode.append(ep_titles[i][3:][:2]) #add episode to episode list
                character.append('Marshall') #add Marshall to character list
                line.append(j.text[10:]) #add following line to line list
            elif 'Lily: ' in j.text: #if line starts with Lily speaking
                season.append(ep_titles[i][:2]) #add season to season list
                episode.append(ep_titles[i][3:][:2]) #add episode to episode list
                character.append('Lily') #add Lily to character list
                line.append(j.text[6:]) #add following line to line list
            elif 'Barney: ' in j.text: #if line starts with Barney speaking
                season.append(ep_titles[i][:2]) #add season to season list
                episode.append(ep_titles[i][3:][:2]) #add episode to episode list
                character.append('Barney') #add Barney to character list
                line.append(j.text[8:]) #add following line to line list
            elif 'Robin: ' in j.text: #if line starts with Robin speaking
                season.append(ep_titles[i][:2]) #add season to season list
                episode.append(ep_titles[i][3:][:2]) #add episode to episode list
                character.append('Robin') #add Robin to character list
                line.append(j.text[7:]) #add following line to line list

while True: #While loop to complete timer
    current_time = time.time() #Setting current time
    elapsed_time = current_time - start_time #Formula to calculate the elapsed time
    
    if elapsed_time > seconds: #if elapsed time is greater than entered number of seconds
        print('Finished iterating in: ' + str(int(elapsed_time)) + ' seconds.') #print elapsed time
        break #break while loop
```

```python
df = pd.DataFrame() #creating df with script information for EDA
df['Season'] = season #setting season to column Season
df['Episode'] = episode #setting episode to column Episode
df['Character'] = character #setting character to column Character
df['Line'] = line #setting line to column Line
df.head() #checking first 5 rows of df
```
